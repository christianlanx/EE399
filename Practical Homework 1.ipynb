### 0.1 Cosine Wave
#### A) Write a function called **StudentCosine** that takes a time array T, frequency f, and amplitude A as arguments, and returns the array X which contains $Acos(2\pi ft)$ for all $t\epsilon T$, using the **numpy** package.

```python
def StudentCosine(T, f, A):
    return A * np.cos(2*np.pi*f*T)
```

#### B) Generate a cosine wave of the length 10000 samples, setting 'T=numpy.arange(0,1,.00001) and call it $x_1$. Set A as 1 and choose $f$ such that you generate 10-20 cycles of the wave.

```python
T = np.arange(0,1,.00001)
A = 1
f = 10
x_1 = StudentCosine(T, f, A)
```

#### C) Plot the waveform $x_1$ versus T using the pyplot module from matplotlib. Also plot $x_1^2$. Report the results.

### 0.3 Fair Coin Toss

```python
def FairCoinTossBOOL():
    return random.random() > 0.5
def FairCoinToss():
    return 'T' if FairCoinTossBOOL() else 'F'
```

#### A) Do 10000 virtual fair coin tosses by calling the function FairCoinToss 10000 times and store the results in a list

```python
fairTosses = [FairCoinToss() for _ in range(10000)]
```

#### B) For $k$ = 10, 20,...,10000, take the first *k* values of the list generated above. Count the number of Heads among the *k* values. Call this number $H(k)$, and calculate the fraction $H(k) / k$. Plot $H(k)/k$ versus *k*. What does $H(k)/k$ converge to? This limit is defined as the probability of Heads. 

```python
k = [x for x in range(10, 10000, 10)]
hK = [len([x for x in fairTosses[:10] if x == 'T'])]
hKoK = [hK[0]/k[0]]
tKoK = [1-hKoK[0]]
for i in range(1, len(k)):
    hK.append(len([x for x in fairTosses[k[i-1]:k[i]] if x == 'T'])+hK[i-1])
    hKoK.append(hK[i]/k[i])
    tKoK.append(1-hKoK[i])
plt.plot(k,hKoK)
```

This value converges to 0.5.

#### C) Repeat part B for Tails for each k and call the number of tails appeared $T(k)/k$. Plot $T(k)/k$ versus $k$. What does $T(k)/k$ converge to? What is this limit called?

```python
plt.plot(k, tKoK)
```

This value also converges to 0.5 and is the probability of tails.

### 0.4 Biased Coin Toss

```python
def BiasedCoinToss(alpha):
    return random.random() > (alpha)
BiasedTosses1 = [BiasedCoinToss(.6) for _ in range(10000)]
BiasedTosses2 = [BiasedCoinToss(.3) for _ in range(10000)]
```

```python
def PrintCoinTossResults(alpha):
    ## 0.5 is a fair toss
    n = 10000
    m = 10
    tosses = [BiasedCoinToss(alpha) for _ in range(n)]
    k = [x for x in range(m, 10000, m)] 
    hK = [len([x for x in tosses[:k[0]] if x])]
    hKoK = [hK[0]/k[0]]
    tKoK = [1-hKoK[0]]
    for i in range(1, len(k)):
        hK.append(len([x for x in tosses[k[i-1]:k[i]] if x])+hK[i-1])
        hKoK.append(hK[i]/k[i])
        tKoK.append(1-hKoK[i])
    
    plt.plot(k, hKoK)
    plt.subplots_adjust(hspace=1, wspace=1)
    plt.subplot(2,1,1)
    plt.title('H(k)/k')
    plt.plot(k, hKoK)
    plt.subplot(2,1,2)
    plt.title('T(k)/k')
    plt.plot(k, tKoK)
```

### 0.5 Eight-sided Die Rolling

#### A) Write a function that virtually rolls an eight-sided die and ouputs the result of the random experiment as strings "1" to "8". Save the function as EightSidedDieRolling

```python
def EightSidedDieRolling():
    r = random.random()
    vals = [1/8, 2/8, 3/8, 4/8, 5/8, 6/8, 7/8, 8/8]
    ## this would be a great place to use a switch statement
    for i in range(len(vals)):
        if (r < vals[i]): return str(i+1)
```

#### B) Do 10000 virtual fair die rollings and store the results in a list.

```python
FairDieRollings = [EightSidedDieRolling() for _ in range(10000)]
```

#### C) For $k$ = 10, 20,...,10000, take the first $k$ values of the list above. Among those $k$ observations, count the number of '1's appeared - call this $1(k)$ - and calculate $1(k)/k$. Plot $1(k)/k$ versus $k$. What does $1(k)/k$ converge to?

```python
oK = [len([x for x in FairDieRollings[:k[0]] if x == '1'])]
oKoK = [oK[0]/k[0]]
for i in range(1, len(k)):
    oK.append(len([x for x in FairDieRollings[k[i-1]:k[i]] if x == '1']) + oK[i-1])
    oKoK.append(oK[i]/k[i])
plt.plot(k, oKoK)
```
